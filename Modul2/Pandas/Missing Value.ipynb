```python
# Isi Missing Value Horsepower dengan Masing Masing Mean dari origin

# Mean Horse Power dari USA
meanUSA = data[data['origin'] == 'usa']['horsepower'].mean()
meanEurope = data[data['origin'] == 'europe']['horsepower'].mean()
```

```python
data.iloc[32] = data.iloc[32].fillna(meanUSA)
```

```python
data.iloc[32]
```

```python
def applyHorsepower(x):
    if(x['horsepower'] > 0):
        return x['horsepower']
    else :
        if(x['origin'] == 'usa'):
            return meanUSA
        else:
            return meanEurope
```

```python
data['horsepower'] = data[[ 'horsepower' ,'origin' ]].apply(applyHorsepower , axis=1)
```

```python
ls = [3,4,2,4,3,4,3,4]
def filterLs (x) :
    return x <= 3
list(filter(filterLs,ls))
```

```python
def duplikatFilter(func,ls):
    newList = []
    for item in ls:
        if(func(item) == True):
            newList.append(item)
    return newList
```

## Titanic Applying Missing Value

```python
dfTitanic = sns.load_dataset('titanic')
```

```python
dfTitanic.info()
```

```python
pd.options.display.max_rows=1000
dfTitanic[dfTitanic['age'].isnull()]
```

```python
import math
```

```python
meanMan = int(dfTitanic[dfTitanic['who'] == 'man']['age'].mean())
meanWoman = int(dfTitanic[dfTitanic['who'] == 'woman']['age'].mean())
meanChild = int(dfTitanic[dfTitanic['who'] == 'child']['age'].mean())
```

```python
def applyAge(x):
    if(x['age'] > 0):
        return x['age']
    else:
        if(x['who'] == 'man'):
            return meanMan
        elif(x['who'] == 'woman'):
            return meanWoman
        else:
            return meanChild
```

```python
dfTitanic['age'] = dfTitanic[['who' , 'age']].apply(applyAge , axis=1)
```

```python
dfTitanic = dfTitanic.dropna(subset=['embarked'])
```

```python
dfTitanic.info()
```

```python
dfTitanic = dfTitanic.drop(columns='deck')
```

```python
dfTitanic.info()
```

```python
dfTitanic= dfTitanic.reset_index(drop=True)
```
