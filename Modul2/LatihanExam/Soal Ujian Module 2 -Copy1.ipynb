### 8. Missing Values
#### Berapa banyak NaN age data dan mean dari setiap who?
#### Isi data NaN dari kolom age dengan mean dari umur tergantung whonya dia apa (child/woman/man)
#### Kemudian print banyak NaN age data dan mean umur setiap who lagi.

```python
ageNa = df[df['age'].isna()]
ageNa
```

```python
ageNa = df[df['age'].isna()]
for item in list(df['who'].unique()):
    totalNan = len(ageNa[ageNa['who'] == item])
    mean = df[df['who'] == item]['age'].mean()
    print('Banyak NaN ' + item + ' & mean : ' + str(totalNan) + ' & ' + str(mean) )
```

```python
def applyAge(x) :
    if(pd.isnull(x['age']) == False):
        return x['age']
    else:
        if(x['who'] == 'man'):
            return int(df[df['who'] == 'man']['age'].mean())
        elif(x['who'] == 'woman'):
            return int(df[df['who'] == 'woman']['age'].mean())
        else:
            return int(df[df['who'] == 'child']['age'].mean())
```

```python
df['age'] = df[['who','age']].apply(applyAge,axis=1)
```

```python
ageNa = df[df['age'].isna()]
for item in list(df['who'].unique()):
    totalNan = len(ageNa[ageNa['who'] == item])
    mean = df[df['who'] == item]['age'].mean()
    print('Banyak NaN ' + item + ' & mean : ' + str(totalNan) + ' & ' + str(mean) )
```
