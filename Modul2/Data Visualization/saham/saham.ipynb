```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# pd.read_csv('TLKM.JK.csv')
df_isat = pd.read_csv('ISAT.JK.csv' , parse_dates=['Date'])
df_fren = pd.read_csv('FREN.JK.csv' , parse_dates=['Date'])
```

```python
df_isat.info()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(df_isat['Date'] , df_isat['Close'] , color='orange' , label='Indosat')
axes.plot(df_fren['Date'] , df_fren['Close'] , color = 'red' , label = 'Fren')
axes.legend()
axes.set_ylabel('Close Price')
axes.set_xlabel('Date / Daily')
```

```python
fig , axes = plt.subplots(nrows=2,ncols=1, figsize=(10,8))
axes[0].plot(df_isat['Date'] , df_isat['Close'],color='orange' , label='Indosat')
axes[1].plot(df_fren['Date'] , df_fren['Close'] , color='red' , label='SmartFren')
axes[0].legend()
axes[1].legend()
axes[0].set_ylabel('Close Price')
axes[0].set_xlabel('Date / Daily')
axes[1].set_ylabel('Close Price')
axes[1].set_xlabel('Date / Daily')
plt.tight_layout()
```
