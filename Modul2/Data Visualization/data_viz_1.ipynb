# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = np.linspace(0,5,11)
```

```python
data2 = data1**2
```

# Functional Matplotlib

```python
data1
```

```python
data2
```

```python
plt.plot(data1,data2,'r-*')
plt.xlabel('Ini X Label')
plt.ylabel('Ini Y Label')
plt.title('Ini Title Dari Plot')
```

### Subplots

```python
#  Parameter = nrow , ncol , plot_number
plt.subplot(1,2,1)
plt.plot(data1,data2)
plt.xlabel('Data 1')
plt.ylabel('Data 2')

plt.subplot(1,2,2)
plt.plot(data2,data1 , 'r')
plt.xlabel('Data 2')
plt.ylabel('Data 1')
```

# OOM Matlotlib

```python
fig = plt.figure()
# add_axes parameter = [left, bottom, width, height]
axes = fig.add_axes([0,0,0.9,0.9])
axes.plot(data1,data2,color='r' , marker = 'o')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,0.9,0.9])
axes2 = fig.add_axes([0.1, 0.1 , 0.3,0.3])
axes1.plot(data1,data2)
axes2.plot(data2,data1)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.15 , 0.7 , 0.1,0.1])
axes3 = fig.add_axes([0.8 , 0.7 , 0.1,0.1])
axes3 = fig.add_axes([0.2 , 0.2 , 0.65,0.1])
axes1.plot(data1,data2)
axes2.plot(data2,data1)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3 , figsize = (10,6))
for item in ax:
    item.plot(data1,data2)
plt.tight_layout()
```

# label, legend, title

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(data1,data2 , color = 'yellow',label='x label')
ax.plot(data1,data1, color = 'red' , label = 'y label')

ax.set_title('Ini Title')
ax.legend(loc=0)
```

# Data Tips

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
# axes.scatter(df['total_bill'], df['tip'])
# axes.set_xlabel('Total Bill')
# axes.set_ylabel('Tip')
# axes.set_title('Total Bill Vs Tip')
axes.scatter(df[df['sex'] == 'Male']['total_bill'] , df[df['sex'] == 'Male']['tip'] , color='red')
axes.scatter(df[df['sex'] == 'Female']['total_bill'] , df[df['sex'] == 'Female']['tip'] , color='blue')
axes.set_ylabel('Tip')
axes.set_xlabel('total_bill')
axes.legend(['Male' , 'Female'])
axes.set_title('Male Vs Female')
```

```python
day = df['day'].unique()
fig = plt.figure()
color = ['red' , 'blue' , 'yellow' , 'green']
axes = fig.add_axes([0,0,1,1])
for i in range(len(day)) :
    axes.scatter(df[df['day'] == day[i]]['total_bill'] , df[df['day'] == day[i]]['tip'] , 
                 color=color[i] , label = day[i])
axes.legend()
axes.set_ylabel('total Bill')
axes.set_xlabel('Tip')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.boxplot(df[df['smoker'] == 'Yes']['tip'],patch_artist=True )
```
