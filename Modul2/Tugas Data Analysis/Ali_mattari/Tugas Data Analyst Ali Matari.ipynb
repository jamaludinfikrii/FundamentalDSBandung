```python
meanFree = df2[df2['Type'] == 'Free']['Rating'].mean()
```

```python
meanFree
```

```python
meanPaid = df2[df2['Type'] == 'Paid']['Rating'].mean()
```

```python
meanPaid
```

```python
def ApplyRating (x):
    if (x['Rating'] > 0):
        return x['Rating']
    else:
        if(x['Type'] == 'Paid'):
            return meanPaid
        else:
            return meanFree        
```

```python
df2['Rating'] = df2[['Rating','Type']].apply(ApplyRating,axis=1)
```
