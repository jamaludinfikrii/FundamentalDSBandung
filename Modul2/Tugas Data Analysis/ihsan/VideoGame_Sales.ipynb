```python
#Import Library Awal
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
#import data frame
game_sales = pd.read_csv("Video_Games.csv")
```

```python
#menampilkan informasi seluruh dataframe
game_sales.info()
```

```python
game_sales
```

```python
#Menampilkan data yang meminili nilai NaN
game_sales.isna().sum()
```

```python
#Melihat percentage data NaN
(game_sales.isna().sum() / len(game_sales) )*100
```

```python
#Walaupun terdapat beberapa data yang memiliki NaN diatas 50 % namun dengan data processing sederhana
#Tanpa mengikuti kaidah clearing data NaN 50 %
#Maka setiap row pada data NaN akan di process sederhana dengan menghapus ROW

game_sales.dropna(axis=0,inplace=True)
```

```python
#sudah tidak ada data NaN data dan data sudah relatif bersih
game_sales.info()
```

```python
game_sales
```

# Tingkat Penjualan Game Untuk Masing - Masing Platform

```python
sns.set_style('whitegrid')
first_fig = plt.figure(figsize=(15,5))
sns.countplot(x=game_sales['Platform'])
plt.tight_layout()
plt.ylabel('Total Penjualan')
```

# Tingkat Penjualan Setiap Region Berdasarkan Game Genre

```python
second_fig = plt.figure(figsize=(15,5))
plt.subplot(121)
plt.pie(
   x=game_sales.groupby('Genre').EU_Sales.sum(),
    labels=game_sales.groupby('Genre').EU_Sales.sum().index,
    shadow=False,
    autopct='%1.1f%%',
    radius = 1.5,
    startangle=45,
    textprops={'fontsize': 11}
)
plt.axis('equal')
plt.title('Europe Genre Sales')

plt.subplot(122)
plt.pie(
   x=game_sales.groupby('Genre').NA_Sales.sum(),
    labels=game_sales.groupby('Genre').NA_Sales.sum().index,
    shadow=False,
    autopct='%1.1f%%',
    radius = 1.5,
    startangle=45,
    textprops={'fontsize': 11}
)
plt.axis('equal')
plt.title('North America Genre Sales')
```

```python
third_fig = plt.figure(figsize=(15,5))
plt.subplot(121)
plt.pie(
   x=game_sales.groupby('Genre').JP_Sales.sum(),
    labels=game_sales.groupby('Genre').JP_Sales.sum().index,
    shadow=False,
    autopct='%1.1f%%',
    radius = 1.5,
    startangle=45,
    textprops={'fontsize': 11}
)
plt.axis('equal')
plt.title('Japan Genre Sales')

plt.subplot(122)
plt.pie(
   x=game_sales.groupby('Genre').Other_Sales.sum(),
    labels=game_sales.groupby('Genre').Other_Sales.sum().index,
    shadow=False,
    autopct='%1.1f%%',
    radius = 1.5,
    startangle=45,
    textprops={'fontsize': 11}
)
plt.axis('equal')
plt.title('Other Region Genre Sales')
```

# Penjualan Global Berdasarkan Age Restriction Category Untuk PS4 vs Xbox

```python
ps4_xbox = game_sales[(game_sales['Platform'] == 'PS4') | (game_sales['Platform'] == 'XOne')]
x = ps4_xbox.groupby(['Rating','Platform']).Global_Sales.sum()
x.unstack().plot(kind='bar',stacked=True,figsize=(15,5))
plt.title('Penjualan Berdasarkan Age Restritction Untuk PS4 vs Xbox')
plt.ylabel('Sales')
```

# Tingkat Korelasi Setiap Row Dan Kolom

```python

fourth_fig = plt.figure(figsize=(15,5))
sns.heatmap(game_sales.corr(), cmap ='gist_earth', linewidth = 0.1, annot=True)
```
