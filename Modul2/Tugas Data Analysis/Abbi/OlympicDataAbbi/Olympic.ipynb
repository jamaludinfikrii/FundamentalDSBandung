```python
meanAge = round(df[df['Sport'] == 'Gymnastics']['Age'].mean())
```

```python
#Mengisi nilai Age dengan mengambil rata2 dari Sex
def ApplyAge(x):
    if (x['Age'] < 0) :
        return x['Age']
    else:
        return meanAge
```

```python
df['Age'] = df[['Sport' , 'Age']].apply(ApplyAge, axis=1)
```

```python
meanHeightM = df[df['Sex'] == 'M']['Height'].mean()
meanHeightF = df[df['Sex'] == 'F']['Height'].mean()
```

```python
#Mengisi nilai Height dengan mengambil rata2 dari Sex
def ApplyHeight(x):
    if(x['Height'] > 0):
        return x['Height']
    else:
        if (x['Sex'] == 'M'):
            return meanHeightM
        else:
            return meanHeightF
```

```python
df['Height'] = df[['Sex' , 'Height']].apply(ApplyHeight, axis=1)
```

```python
df.info()
```

```python
meanWeightM = df[df['Sex'] == 'M']['Weight'].mean()
meanWeightF = df[df['Sex'] == 'F']['Weight'].mean()
```

```python
#Mengisi nilai Weight dengan mengambil rata2 dari Sex
def ApplyWeight(x):
    if(x['Weight'] > 0):
        return x['Weight']
    else:
        if (x['Sex'] == 'M'):
            return meanWeightM
        else:
            return meanWeightF
```

```python
df['Weight'] = round(df[['Sex' , 'Weight']].apply(ApplyWeight, axis=1))
```
