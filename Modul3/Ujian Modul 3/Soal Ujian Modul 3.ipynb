## 9. Use K Fold (K=5,shuffle=True,random_state=42), every loop use RandomOverSampler for the training data, and then use GradientBoostingClassifier for the model

```python
from sklearn.metrics import accuracy_score,roc_auc_score,f1_score,matthews_corrcoef,log_loss,classification_report
```

```python
def all_metrics_report(X_train,X_test,y_train,y_test,model):
    predict_train = model.predict(X_train)
    predict_test = model.predict(X_test)
    predict_proba_test = model.predict_proba(X_test)
    predict_proba_train = model.predict_proba(X_train)
    train_test_score = {
        'Train Accuracy' : accuracy_score(y_train,predict_train),
        'Test Accuracy' : accuracy_score(y_test,predict_test),
        'Train ROC AUC' : roc_auc_score(y_train,predict_proba_train[:,1]),
        'Test ROC AUC' : roc_auc_score(y_test,predict_proba_test[:,1]),
        'Train F1 Score' : f1_score(y_train,predict_train),
        'Test F1 Score' : f1_score(y_test,predict_test),
        'Train Matthews Corr Coef' : matthews_corrcoef(y_train,predict_train),
        'Test Matthews Corr Coef' : matthews_corrcoef(y_test,predict_test),
        'Train Log Loss' : log_loss(y_train,predict_proba_train),
        'Test Log Loss' : log_loss(y_test,predict_proba_test),  
    }
    
    class_report_train = classification_report(y_train,predict_train)
    class_report_test = classification_report(y_test,predict_test)
    
    return train_test_score,class_report_train,class_report_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test,y_train,y_test = train_test_split(x_pca_df,df["Class"])
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
score,cf_train,cf_test = all_metrics_report(X_train,X_test,y_train,y_test,model)
```

## 10. Print the Model Evaluation like below from the K Fold 

```python
from sklearn.model_selection import KFold
from imblearn.over_sampling import RandomOverSampler
```

```python
kf = KFold(n_splits=5,shuffle=True,random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
data_metric = []
class_report = []
for train_i,test_i in kf.split(x_pca_df,df['Class']):
    X_train,X_test,y_train,y_test = x_pca_df.iloc[train_i] , x_pca_df.iloc[test_i], 
    df['Class'].iloc[train_i] ,df['Class'].iloc[test_i]
    ros = RandomOverSampler()
    X_ros,y_ros  = ros.fit_sample(X_train,y_train)
    model = GradientBoostingClassifier()
    model.fit(X_ros,y_ros)
    score,class_train,class_test = all_metrics_report(X_ros,X_test,y_ros,y_test,model)
    data_metric.append(score)
    class_report.append([class_train,class_test])
```

```python
df_metrics_score = pd.DataFrame(data_metric)[['Train Accuracy','Test Accuracy','Train ROC AUC','Test ROC AUC','Train F1 Score',
                           'Test F1 Score','Train Matthews Corr Coef','Test Matthews Corr Coef','Train Log Loss','Test Log Loss']]
```

```python
for col in df_metrics_score[['Train Accuracy','Test Accuracy','Train ROC AUC','Test ROC AUC']].columns:
    print('Avg ' + col + ' = ' + str(df_metrics_score[col].mean()))
```
